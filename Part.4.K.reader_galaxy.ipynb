# Galaxy: 用作品社交，成为一群显赫之人的一员

——` 心存梦想，追随兴趣，邂逅无尽可能……`or` 不断地自我否定，是进步还是退步？`

Hello，大家好，我是**Galaxy**。很荣幸能有机会跟大家分享我的个人成长故事，希望能通过我自己的成长经历给大家提供一些可参考的思路，起到抛砖引玉的效果。回首来路，我好像是个 “善变” 且不断 “自我否定” 的人……

## Galaxy 词解：一群显赫的人

首先，我要像大家隆重介绍一下我的英文名和网络昵称，**Galaxy**，这个单词有三层含义。最为大家所熟知的第一层含义应该是：三星的一款智能手机品牌，注册于 2015 年；第二层含义是：星系、银河系的意思；第三层含义比较少为人知却是我最喜欢的一层含义：**一群显赫的人**。

我特别喜欢这个词，喜欢 “**一群显赫的人**” 这层意义。**显赫**，代表了不甘平庸，**一群**，有呼朋引伴的意思。与一群有共同梦想的人一起共事、一起成长，是我一直的梦想与努力方向…… Galaxy 的寓意精准地承载了我的梦想，所以我对这个词一直情有独钟。从 2008 年入职外企我就一直使用这个英文名，直到现在，我的网络昵称始终都是 Galaxy。

人人都有一颗不甘平庸的心，但是只有愿望还不够，**要用执着的行动走出一条显赫之路**。在我的个人成长之路上，每一次重大选择都是我自己拿的主意，所以那些选择对我来说都是 “最好” 的。所以，貌似每一次最终都成功了。但这并不意味着那些目标都是轻而易举、一帆风顺就能实现的。

## 我的经历：成长如生活永无终点

我研究生毕业后直接进了一家**外企**，后来辞职**回归家庭**，现在又活跃于**网上社区**……

### 我的外企情缘

我从小就是个兴趣驱动型人。中学时特别喜欢学数学和物理。然后高考报志愿的时候就非要报物理学，当时根本就不考虑就业方向什么的。我最不愿意从事的职业就是教师和医生。但是很讽刺的是，我当时的成绩上陕西师范大学是最划算的，后来被亲人说服去了师大，当然是物理专业。家人建议报化学、英语、计算机什么的，说那些专业简单，我说：“物理难怎么了，我喜欢学，就能搞定”。

当时在中学的时候应该没怎么见过计算机，对计算机科学专业也没有概念。所以，更讽刺的是，我刚进大学没多久，就发现自己兴趣转移了，朦胧地觉得计算机知识很奇妙，很想将来从事编程行业。然后我就暗下决心，自学计算机知识，打算考研的时候转专业，以后要进一家外企工作。

那会儿我的主要教材就是那本清华大学出版的 C 语言，就那一本书，反复翻、反复看、反复画，也自学了计算机领域其他的一些相关知识。因为物理学跟计算机专业交叉实在太少，为了平滑过渡，我考了西北工业大学的电路与系统专业，离计算机科学进了一步。然后利用实习期，自学 C++并实际参与了一些项目，就因为这段宝贵的实习经历，以及我自身的兴趣驱动型人格，后来同时拿到北京和西安两家外企的 Offer，最终决定留在西安，就放弃了北京的机会。

成功的那一刻，感觉梦想那么容易就实现了呢！所以，只要有追求、有梦想，目标明确，就一定能想方设法去实现。

### 这么费劲进了外企，怎么舍得回归家庭呢？

没有孩子之前，我也特别不理解为什么很多妈妈有了孩子就要辞职呢！只有亲身经历那个角色才能感同身受，孩子那么小的时候，就是很依恋妈妈。我也清楚地知道对于孩子的成长我能起到无可替代的作用。其次，随着身份的转变，我的兴趣又一次转移，我对各种非传统的教育理念很着迷。再次，回归不是牺牲，只是**选择了一条不一样的成长路径**。一条可以兼顾梦想，把孩子一起规划进来的成长之路。

其实，很多人认为回归家庭就是脱离社会，就是放弃自我。我可以很负责任地说，脱离企业才是海阔天空，才有更多可能。具体的工作岗位对于一个人多样化的潜力其实是一种限制，你的很多价值其实没有机会发挥出来，你的社交圈也是有局限的。回归家庭，以前的同事可能慢慢没了共同语言，但现在是网络时代啊，你走出那个小圈子，将自己暴露在更广阔的网络世界中，其实是有更多机会链接到志同道合的伙伴。

世界那么大，相同的兴趣就是最好的粘合剂。那个阶段我主要关注亲子教育领域，自己研读经典书籍，加上实际的育儿经验，喜欢思考总结，偶尔写写公众号。后来因为一篇文章，找到了组织，参加并自己主持亲子读书会，结识了这个领域最优秀的一波朋友，而且有幸被邀请加入核心团队。

那段时间，研读了很多经典书籍，包括教育、亲子、心理学等领域，不知不觉中完成了很重要的积累过程。

### 现在为什么又转战网络社区了呢？

优秀的群体都有交集。我偶然从导读师群里看到笑来老师的文章，然后就走上了个人认知升级之路。孩子慢慢长大，兴趣也再一次与时俱进了。

不甘平庸，就要在哪里都积极争取机会，创造更多可能性。我现在主要活跃于 PRS 托荒者社区和 Mixin 社区。即便没有经验，但是我有热情和挑战自我的信心，我有执着实现目标的韧性和自驱力，我有停不住的思考和不知疲惫的大脑。拥有这些积极的特质，让自己成为自带正能量的人，必定会成就 “显赫” 的结局。

在拓荒者社区，我参与社群管理、运营新媒体账号、**参与产品 idea 创意大赛并意外获奖**，自己主导组建超级节点等等。如果不是不甘平庸，如果不是一路追随梦想，渴望成长，这些不可能跟我有关系的。在 Mixin 社区，从来没有销售经验，但是践行群的销售业绩还不错；签到活动也是社区里为数不多的一直超前完成任务的人之一。

记得产品 idea 征集活动刚启动的时候，因为我是社区积极分子，所有活动都不会 “放过”。但是，说实话，我当时想的就是，这么专业的事，估计也只能凑凑热闹，只当增添人气了。但是决定要做的事，就算不自量力，也至少要努力到一百分。我就积极思考、琢磨、在脑海中搜寻素材、用搜索引擎临时补课，虽然没有任何基础，虽然完全不懂产品，没想到最终居然一路过关斩将，意外地进入获奖 idea 行列。

## 我的做法：如何成为 “一群” 一员

我不甘平庸，却也不喜欢单打独斗，我向往与一群显赫的人一起做有意思的事。

### 我是如何让吸引力法则在自己身上起作用的？

兴趣引发行动，行动形成积累，积累转化为作品。然后，用作品社交，触达 “同类”。所谓作品不一定要很宏伟。一个个最小化作品也一样有力量。比如，文章、评论、心得、内化总结，这些形式可以交叉使用。

- 在亲子教育领域，也许就是因为那篇 [《没有利益刺激的好评背后 -- 是从我们这群人身上能闻到的一样的味道！》](https://mp.weixin.qq.com/s/oNMj1DpDBv6OcK3uyqHFPA)，让我有幸加入核心团队，跟一群优秀的人共事。

- 在拓荒者社区，也许就是因为那个 [《桥段汇 —— 为内容增色，为创作助力》](https://mp.weixin.qq.com/s/Ze4Se-_6JfuBa6AqfVAB2g) 的获奖产品 Idea，为我奠定了信任基础，如今才能这么顺利的组建一个优秀的超级节点团队。

- 在 Mixin 社区，就是因为我的一篇篇文章、一篇篇内化总结，帮我聚集了一个特别温馨有爱的小社群。通过作品双向筛选聚到一起的朋友，三观合得来，有着部分交叠的认知背景，一起交流学习，很亲切，有那种所谓强链接带来的幸福感。

之前我都是由于自己的 “显赫” 被邀请加入核心团队，这已经让我不胜欣喜了，幸福感满满。自己主导组建一个团队当然不是没想过，但没想到来的这么快、这么突然。PRS 超级节点我本想着够资格参与就不错了，都没敢想过成为团队一员，可是惊喜就是来的这么 “猝不及防”。因为用作品社交积累的人气与信任，我居然第一次站在了小小 leader 的位置。我们会努力、我们会珍惜、我们会一直一起走……

### 什么样的作品可以称之为 “作品”？

其实，人人都有积累，多少不同，领域不同罢了。偶尔灵感涌现，写一篇脍炙人口的文章对谁来说都不是什么难事。难的是，可持续。本来就没多少积累，还想要日更，还要求篇幅、质量，这个门槛确实不低。

积累不够，想要有高质量的输出，那感觉就是举步维艰。积累达不到某个阈值，就谈不上融会贯通，也就很难有独立见解，就不可能发现新的主题，自然无法摆脱路径依赖。**内化总结**是我找到的一个**完美缓冲策略**。首先，总结是篇幅越短越好，用尽量精炼的文字传达最丰富、最深刻的内涵；其次，既然是总结，那就自带主题，从自创主题变成寻找、筛选主题，难度降了一个量级；最后，保质保量的输出内化总结，顺便也实现了高质量的积累。

以微博为例，输出形式无非以下四种：

- 头条文章
- 转发微博

头条文章，无论是自己写的还是转发别人的，都需要附加一段自己的总结，给读者一个是否有必要阅读全文的参考，因此和转发微博类似。直接转发微博毫无技术含量，自然也不大可能产生任何积累。让自己的每一条输出，无论长短，都要有分量，都能引发共鸣，否则意义不大。

- 关注的动态

我现在的注意力主要就放在 PRS 和 Mixin 社区，我的活跃阵地就在这些地方。所以，类似 Mixin 签到活动、新功能上线、xue.cn 的各种活动、PRS 超级节点等等，都会同步过来，一边分享心情一边传递信息。

- 常规分享

于我就是每日内化总结。这半年多，除了践行群课程，我总结了 [两本整书](https://weibo.com/ttarticle/p/show?id=2309404457077283095012)，第三本马上也完了。还有一些著名的演讲稿，甚至有时在群里跟伙伴们聊天碰撞出灵感，也能形成特别深刻的总结。所有能触动自己的内容，都值得做内化总结。

前三类分享算是一种调剂，可以丰富内容、活跃气氛，可遇不可求。因此，需要有一项常规输出。我做不到每天写上千字的文章是事实。但是，我也发现，即使你能写出来，即使你水平也比较高，大家也没有那么多精力每天看头条文章。而简短凝练的内化总结，一览无余，不会给人造成阅读压力。

在此基础上，如果能引发共鸣、甚至震撼，如果能够提供价值、刷新认知，那就是非常成功的日常输出了。所以，我定义的内化总结是这样的：

> 用尽量短的篇幅表达一个逻辑严谨且完整的内容，想要触动心灵，就要反复推敲、字斟句酌，到最后，甚至做到句句金句，而且大都是原创的。

### 积累不够怎么解决呢？

不仅要**多输入**，还必须是**有效输入**。不然看了很多书，完了什么也没记住，那还是没有积累。我个人觉得，我经历的两个阶段的书籍解读对自己的积累很有帮助。第一个阶段，在保证逻辑完整的前提下，反复阅读、反复划重点、一遍一遍删减，**主要任务是浓缩**，提炼出一本书最精华的内容，不参杂自己的主观解读。第二个阶段，**解读就是内化总结**，依然反复阅读、反复划重点，然后启用元认知能力揣摩作者的思路，梳理逻辑架构，最终以自己的见解重组核心观点，输出有自己阅历烙印的总结。

### 如何做到自驱动，乐此不疲地输出 “作品”？

内化总结，绝对不是随随便便、轻轻松松的事情，但是它真真切切带给我很多欢乐。每当完成一份自己特别满意的 “作品” 时，我都非常开心，得意忘形地开心，又蹦又跳的自嗨。不说这项技能会带给我一些潜在经济收益的可能；会帮我建立物理世界无法想象的有效社交；仅从一件作品至少收获一份自发的快乐，从细微处提高生活品质来说，就相当值得。活着，开心多重要，这难道不足以成为一个长期的目标和动力吗？

心存梦想，追随兴趣，该任性的时候就任性到底。将就和委屈求全从来都不在我的选择之列，哪怕后来证明选错了，也无怨无悔，大不了迂回改道，无非是不一样的精彩！生命不息，折腾不止，自学不停，输出不断，加油！
